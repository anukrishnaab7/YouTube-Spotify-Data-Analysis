# Data Analysis

```python
data.head(10)
```

# Top 10 artists on the Spotify platform whose songs have more views.

```python
Top_artist= data.groupby('Artist')['Views'].agg('sum').reset_index().sort_values('Views', ascending=False).head(10)

print('Top Artist based on Views: ','\n','\n', Top_artist)
```

```python
Top_artist.set_index('Artist', inplace=True)
Top_artist.plot(kind= 'bar', color='g')

plt.xlabel('Artist')
plt.ylabel('Views')
plt.title('Top 10 Artist with higher views')
plt.xticks(rotation=80)
plt.show()
```

# Top 10 artist with most Likes on youtube

```python
Top_artist_1= data.groupby('Artist')['Likes'].agg('sum').reset_index().sort_values('Likes', ascending=False).head(10)
print('Top_artist based on likes: ', '\n','\n', Top_artist_1)
```

```python
ax=data.groupby('Artist').agg({'Likes':'sum','Stream':'sum'}).reset_index()
bx=ax.sort_values('Likes',ascending=False).head(10)

sns.barplot(data=bx,x='Likes',y='Artist')
plt.title('Top 10 most Liked Artist ')
plt.show()
```

# Determine the top 10 Streamers on spotify


```python
cx=ax.sort_values('Stream',ascending=False).head(10)
cx_1= data.groupby('Artist')['Stream'].agg('sum').head(10)

print('Top 10 Streamers on spotify: ','\n','\n', cx_1)
```

```python
sns.barplot(data=cx,y='Stream',x='Artist')
plt.xticks(rotation=80)
plt.title('Top 10 streamers')
plt.show()
```

# Most commented artist on youtube


```python
ax=data.groupby('Artist')['Comments'].sum().reset_index().sort_values(by='Comments',ascending=False).head(10)
print('Most Commented artist on youtube: ', '\n', ax)
```

```python
sns.barplot(data=ax,y='Artist',x='Comments')
plt.title('Top 10 most commented Artist')
plt.show()
```
